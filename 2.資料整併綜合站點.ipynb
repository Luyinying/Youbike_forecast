```python
def insert_time_W(row):
  return row['時間'].replace(year=2021)
Weather['時間'] = Weather.apply(lambda r: insert_time_W(r), axis=1)
Weather = Weather.reset_index(drop=True)
Weather
```

```python
def weather_mix(D,W):
  K = D.merge(W,how='inner')
  return K
```

```python
def MRT_add(a,kk):
  D = []
  for i in range(len(kk)):
    D.append(a)
  D = pd.DataFrame(D,columns=['捷運站'])
  kk = pd.concat([kk,D],axis=1)
  return kk
```

```python
def add_mrt(g1,h1):
  g1['時間'] = pd.to_datetime(g1['時間'])
  h1['時間'] = pd.to_datetime(h1['時間'])
  a1 = g1.merge(h1,how='inner')
  a1 = a1.reset_index(drop=True)
  a2 = g1.merge(h1,how='outer',indicator=True).loc[lambda x : x['_merge'] == 'left_only']
  a2 = a2.drop('_merge',axis=1)
  a2['出站人數'] = 	a2['出站人數'].fillna(0)
  a2['進站人數'] = 	a2['進站人數'].fillna(0)

  aa = pd.concat([a1,a2],axis=0)
  aa['時間'] = pd.to_datetime(aa['時間'])
  aa = aa.sort_values(by=['時間'])
  aa = aa.reset_index(drop=True)
  return aa
```
