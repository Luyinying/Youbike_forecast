```python
import pandas as pd
```

```python
ticket01 = pd.read_csv('/content/drive/MyDrive/研究所課程/期末專題_捷運/data/ubike_ticket_001.csv',encoding='latin1',header=None)
ticket02 = pd.read_csv('/content/drive/MyDrive/研究所課程/期末專題_捷運/data/ubike_ticket_002.csv', encoding='latin1',header=None)
ticket03 = pd.read_csv('/content/drive/MyDrive/研究所課程/期末專題_捷運/data/ubike_ticket_003.csv', encoding='latin1',header=None)
data = ticket01.append(ticket02)
data = data.append(ticket03)
data = data.reset_index(drop=True)
```

```python
data.columns=['交易序號','卡號','扣款時間','扣款金額','借車時間','借車站代號','借車站','借車車柱','自行車編號','還車時間','還車站代號','還車站','還車車柱','租用時間','手機序號','卡片種類','費率別','業者代碼','根據檔案是IPASS或ECC','整批資料更新時間','存入DB時機','該筆資料更新時間','該筆資料更新日期']
```

```python
data.isnull().sum()
```

```python
data = data.dropna(axis=0, how='any')
```

```python
data.isnull().sum()
```

```python
data = data.drop('交易序號',axis=1)
data = data.drop('卡號',axis=1)
data = data.drop('扣款時間',axis=1)
data = data.drop('卡片種類',axis=1)
data = data.drop('業者代碼',axis=1)
data = data.drop('根據檔案是IPASS或ECC',axis=1)
data = data.drop('整批資料更新時間',axis=1)
data = data.drop('存入DB時機',axis=1)
data = data.drop('該筆資料更新時間',axis=1)
data = data.drop('該筆資料更新日期',axis=1)
```

```python
df = data
```

```python
df['借車時間'] = pd.to_datetime(df['借車時間'])
```

```python
df = df.sort_values(by=['借車時間'])
```

```python
df['還車時間'] = pd.to_datetime(df['還車時間'])
```

```python
mask = df[(df['借車時間'] >= '2017-07-01 00:00:00') & (df['借車時間'] <= '2017-07-31 23:59:59')]
mask = mask.reset_index(drop=True)
mask
```

```python
def insert_time(row):
  return row['借車時間'].replace(year=2021)
mask['借車時間'] = mask.apply(lambda r: insert_time(r), axis=1)
```

```python
def insert_time1(row):
  return row['還車時間'].replace(year=2021)
mask['還車時間'] = mask.apply(lambda r: insert_time1(r), axis=1)
```

```python
aa = mask[(mask['還車時間']-mask['借車時間']<='08:00:00') & (mask['還車時間']-mask['借車時間']>='00:05:00')]
aa = aa.reset_index(drop=True)
aa
```

```python
LL = pd.DataFrame(aa['借車站'])
a = []
for col in LL.columns:
    if LL[col].dtypes == 'object':
      a.append(LL[col].unique())
a = pd.DataFrame(a).T
a.columns=['站點']
```

```python
p = []
for i in range(len(a)):
  n = a['站點'][i]
  new= n.encode('latin-1').decode('utf-8')
  p.append(new)
  print(new)
```

```python
import numpy as np
g = Le.groupby(Le['借車站'])
count = []
for name, group in g:
  a = group.groupby(Le['借車時間'].dt.date)
  for name2, group2 in a:
    b = group2.groupby(Le['借車時間'].dt.hour)
    for name3, group3 in b:
      count.append([str(name), str(name2)+' '+str(name3)+ ':00:00',len(group3)])
count
```

```python
df_le = pd.DataFrame(count)
df_le.columns =['站點','時間','借車數']
df_le
```

```python
df_le['時間'] = pd.to_datetime(df_le['時間'])
df_le
```
